```python
rng = np.random.RandomState(7)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.randn(100)
sizes = 1000*rng.rand(100)

plt.scatter(x,y, c=colors, s=sizes, alpha=0.5, cmap='Reds')
plt.colorbar()
```

```python
x = np.random.normal(0, 0.8, 1000)
y = np.random.normal(-2, 1, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.5, normed=True, bins=50)

plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```

```python
mean = [5,2]
cov = [[3,1],[2,0.5]]

x,y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=20, cmap= 'Reds')
cb = plt.colorbar()
```

```python
plt.hexbin(x, y, gridsize=20, cmap='Reds')

cb = plt.colorbar()
```
